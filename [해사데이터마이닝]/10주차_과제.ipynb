```python
문제 3) 일원분산분석: 지점별 고객만족도 비교

모범답안
Q1. 가설 설정 귀무가설 (H_0): 세 지점의 평균 만족도는 모두 같다. 대립가설 (H_1): 적어도 한 지점의 평균 만족도는 다르다.
Q2. 검정 방법 선택 3개 이상 집단 평균 비교이므로 일원 ANOVA를 사용한다.
Q3. 검정 수행 지점별 데이터를 나누어 f_oneway()를 수행한다.
Q4. 평균 비교 각 지점 평균을 계산해 가장 높은 지점을 확인한다.
Q5. 결론 p-value < 0.05이면 지점 간 만족도 차이가 존재한다.

Python 예시 코드 import pandas as pd from scipy import stats df = pd.read_csv("branch_satisfaction.csv") a = df[df["branch"] == "A"]["score"] b = df[df["branch"] == "B"]["score"] c = df[df["branch"] == "C"]["score"] result = stats.f_oneway(a, b, c) means = df.groupby("branch")["score"].mean()

print("지점별 평균:") print(means) print("F-statistic:", result.statistic) print("p-value:", result.pvalue) print("평균이 가장 높은 지점:", means.idxmax())

핵심 결과:
• A지점 평균: 76.3773
• B지점 평균: 82.9443
• C지점 평균: 74.3860
 • F-statistic: 18.0787 • p-value: 0.000000
  • 평균 최고 지점: B지점 결과에 대한 해석
  :일원분산분석을 실시한 결과, 지점별 평균 고객만족도는 A지점 76.3773점, B지점 82.9443점, C지점 74.3860점으로 나타났다. 검정 결과 F=18.0787, p-value는 2.72e-07로 유의수준 0.05보다 작으므로, 지점 간 평균 고객만족도는 동일하지 않으며 통계적으로 유의한 차이가 존재한다고 해석할 수 있다. 특히 세 지점 중에서는 B지점의 평균 만족도가 가장 높게 나타났다.

최종 해석 유의한 경우:“일원분산분석 결과 p-value가 0.05보다 작으므로, 세 지점의 평균 고객만족도는 동일하지 않으며 적어도 한 지점은 다른 지점과 차이가 있다.” 유의하지 않은 경우:“일원분산분석 결과 p-value가 0.05 이상이므로, 지점 간 평균 만족도 차이가 있다고 보기 어렵다.” 일원 ANOVA는 SciPy의 f_oneway로 수행할 수 있습니다.



```

```python
문제 4) 이원분산분석: 광고유형과 연령대가 구매금액에 미치는 영향

문제 제시문 마케팅팀은 광고유형(영상광고, 이미지광고)과 연령대(20대, 30대, 40대)가 구매금액에 미치는 영향을 동시에 분석하려고 한다.특히 광고유형의 효과가 연령대에 따라 달라지는지 확인하고자 한다.

모범답안
Q1. 가설 설정 광고유형 주효과: 광고유형에 따른 평균 구매금액 차이가 없다. 연령대 주효과: 연령대에 따른 평균 구매금액 차이가 없다. 상호작용 효과: 광고유형과 연령대의 상호작용 효과가 없다.
Q2. 검정 방법 선택 독립변수 2개, 종속변수 1개이므로 이원 ANOVA를 사용한다.
Q3. 광고유형 주효과 판단 ANOVA 표에서 C(ad_type)의 p-value 확인
Q4. 연령대 주효과 판단 C(age_group)의 p-value 확인
Q5. 상호작용 해석 C(ad_type):C(age_group)의 p-value가 유의하면 연령대에 따라 광고유형 효과가 달라진다.

Python 예시 코드 import pandas as pd from scipy import stats df = pd.read_csv("before_after_worktime.csv") before = df["before_min"] after = df["after_min"] diff = before - after result = stats.ttest_rel(before, after, alternative="greater")
print("평균 차이(before - after):", diff.mean()) print("t-statistic:", result.statistic) print("p-value:", result.pvalue)

핵심 결과:- 광고유형 p-value: 4.81e-13 - 연령대 p-value: 3.09e-09 - 상호작용 p-value: 0.002236 결과에 대한 해석:이원분산분석 결과, 광고유형의 p-value는 4.81e-13, 연령대의 p-value는 3.09e-09, 광고유형과 연령대의 상호작용 p-value는 0.002236으로 모두 유의수준 0.05보다 작게 나타났다. 따라서 광고유형과 연령대는 각각 구매금액에 유의한 영향을 미치며, 동시에 광고유형의 효과는 연령대에 따라 달라지는 상호작용 효과도 존재한다고 해석할 수 있다. 즉, 동일한 광고유형이라도 연령대별 반응 차이를 고려한 마케팅 전략이 필요하다.
최종 해석
 • 광고유형 유의:“광고유형의 p-value가 0.05보다 작으므로 광고유형에 따라 평균 구매금액 차이가 있다고 볼 수 있다.”
 • 연령대 유의:“연령대의 p-value가 0.05보다 작으므로 연령대에 따라 평균 구매금액 차이가 있다고 볼 수 있다.”
  • 상호작용 유의:“상호작용 p-value가 0.05보다 작으므로 광고유형의 효과는 연령대에 따라 달라진다고 해석할 수 있다.”
   R-style formula 기반 OLS와 분산분석은 statsmodels 문서 흐름에 맞게 사용할 수 있습니다.

```
