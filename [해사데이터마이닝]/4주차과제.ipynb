```python
[문제9] air_quality.csv 활용문제
IQR을 이용해 이산화탄소(CO2) 이상치 수를 찾으시오.

-Q1(제1사분위수): 하위 25%, Q3(제3사분위수): 50%~75%
-IQR = Q3 – Q1
-lower(하한선): Q1 – 1.5 IQR, 이 값보다 작은 데이터는 이상치
-upper(상한선): Q3 + 1.5 IQR, 이 값보다 큰 데이터는 이상치

[문제9-Ans]
import pandas as pd
# df = pd.read_csv("air_quality.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch7/air_quality.csv")

# 1) IQR 계산
Q1 = df['CO2'].quantile(0.25)
Q3 = df['CO2'].quantile(0.75)
IQR = Q3 - Q1

# 2) 상한 및 하한 계산
upper = Q3 + 1.5 * IQR
lower = Q1 - 1.5 * IQR
# 3) 이상치 식별
outliers = df[(df['CO2'] < lower) | (df['CO2'] > upper)]

# 4) 이상치 수
print(len(outliers))

정답: 304
```

```python
[문제12] chem.csv 활용문제
1) 주어진 데이터에서 ‘co’와 ‘nmhc’ 컬럼을 각각 Min-Max 스케일링 하시오.
2) 스케일링된 ‘co’, ‘nmhc’ 컬럼의 표준편자를 각각 구하시오.
3) ‘co’ 컬럼으 표준편차에서 ‘nmhc’ 컬럼의 표준편차를 뺀 값을 소수점 3자리로 반올림하여 구하시오.

[문제12-Ans]
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# df = pd.read_csv("chem.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch8/chem.csv")

# 1) Min-Max 스케일링 (데이터프레임으로 만들기)
scaler = MinMaxScaler()
df['co_scaled'] = scaler.fit_transform(df[['co']])
df['nmhc_scaled'] = scaler.fit_transform(df[['nmhc']])

[문제12-Ans]
# 2) 표준편차 계산 (판다스 표준편차)
co_std = df['co_scaled'].std()
nmhc_std = df['nmhc_scaled'].std()
print(co_std, nmhc_std)
정답: 0.2856516497116944 0.3030617020578397

# 3) 표준편차 차이 계산 및 반올림
std_diff = round(co_std - nmhc_std, 3)
print(std_diff)
정답: -0.017
```

```python
[문제14] crime.csv 활용문제
1)  "발생건수"와 "검거건수"를 따로 분리 하시오.
2)  검거율 계산 (검거건수 / 발생건수) 하시오.
3)  각 연도에서 검거율이 가장 높은 범죄유형 찾으시오.
4) 가장 높은 검거율을 기록한 범죄유형의 검거건수를 구하시오.

[문제14-Ans]
import pandas as pd
# df = pd.read_csv("crime.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert_v2/main/part4/ch9/crime.csv")

# 1) "발생건수"와 "검거건수"를 따로 분리
cond1 = df["구분"] == "발생건수"
cond2 = df["구분"] == "검거건수"
df1 = df[cond1].iloc[:, 2:] # 발생건수만 가져오기 df2 = df[cond2].iloc[:, 2:] # 검거건수만 가져오기

#[:, 2:] 슬라이싱 의미 : 행은 모두, 열을 2번 위치부터 끝까지 가져오라는 뜻

# 2) 검거율 계산 (검거건수 / 발생건수)
df1 = df1.reset_index(drop=True)
df2 = df2.reset_index(drop=True)
df3 = df2 / df1
#reset_index(drop=True)는 필터링 후 남아 있는 행들의 인덱스를 0, 1, 2, ...처럼 다시 새로 매기는 작업. 그리고 drop=True는 기존 인덱스를 새로운 열로 남기지 않고 버린다.

# 3) 각 연도에서 검거율이 가장 높은 범죄유형 찾기 listbox = df3.idxmax(axis=1)
#df3의 각 행에서 가장 큰 값의 열 이름을 구합니다.

# 4) 가장 높은 검거율을 기록한 범죄유형의 검거건수 가져오기
result = 0
for index, item in enumerate(listbox):
     result = result + df2.loc[index, item]
print(result)

정답: 7799
```

```python
[문제15] hr.csv 활용문제
1)  만족도 결측치를 평균으로 처리 하시오.
2) 근속연수 결측치 처리 (부서와 성과등급 기준 평균값으로 채움) 하시오.
3)  연봉 / 근속연수 계산 후 세 번째로 높은 사람의 근속연수를 구하시오.
4) 연봉 / 만족도 계산 후 두 번째로 높은 사람의 교육참가횟수를 구하시오.
5) 3)+4)의 결과 출력하시오.

[문제15-Ans]
import pandas as pd
# df = pd.read_csv(“hr.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert_v2/main/part4/ch9/hr.csv")
# 1) 만족도 결측치 처리
m = df["만족도"].mean()
df["만족도"] = df["만족도"].fillna(m)

# 2) 근속연수 결측치 처리 (부서와 성과등급 기준 평균값으로 채움)
gm = df.groupby(["부서", "성과등급"])["근속연수"].transform("mean")
#근속연수는 전체 평균보다 같은 부서, 같은 성과등급 사람들의 평균으로 채우는 것이 더 자연스럽다

# 3) 연봉 / 근속연수 계산 후 세 번째로 높은 사람의 근속연수 (A)
df["연봉_근속연수"] = df["연봉"] / df["근속연수"]
# “근속연수 1년당 연봉이 얼마나 되는가” 같은 비율값 생성 예) 연봉 = 6000, 근속연수 = 3 이면, "연봉_근속연수" 값은 2000

# 4) 연봉 / 만족도 계산 후 두 번째로 높은 사람의 교육참가횟수 (B)
df["연봉_만족도"] = df["연봉"] / df["만족도"]
df_like = df.nlargest(2,'연봉_만족도’)

# "연봉_만족도" 열을 기준으로 값이 큰 순서대로 상위 2개 행을 뽑는다.
B = df_like.iloc[-1]["교육참가횟수"]
# 5) 결과 출력
result = A + B
print(result)

정답: 7.0
```

```python
1. pandas.melt()
역할: 데이터의 형태를 '넓은 형태(Wide)'에서 '긴 형태(Long)'로 재구조화(Reshaping)
기능: 표의 열(Column)로 나란히 퍼져있는 데이터를 행(Row)으로 길게 녹여냅니다(melt).
언제 쓰나요? 예를 들어 '2021년 매출', '2022년 매출', '2023년 매출'이 각각의 열로 되어있는 표를, '연도'라는 하나의 열과 '매출'이라는 하나의 열로 합쳐서 데이터를 길게 만들고 싶을 때 사용합니다. 데이터를 시각화하거나 데이터베이스에 저장하기 좋은 형태로 바꿀 때 필수적입니다.

2. DataFrame.groupby()
역할: 데이터를 특정 기준에 따라 그룹화 (Split-Apply-Combine)
기능: 지정한 열의 고유한 값들을 기준으로 데이터를 여러 그룹으로 쪼갭니다.
언제 쓰나요? 전체 학생의 성적 데이터가 있을 때, "반별로(기준)" 수학 점수의 "평균(연산)"을 구하고 싶을 때 사용합니다. 데이터를 그룹으로 묶은 뒤 합계, 평균, 최댓값 등의 통계량을 구할 때 무조건 쓰이는 메서드입니다.

3. DataFrame.sort_values()
역할: 데이터 정렬
기능: 지정한 특정 열(또는 여러 열)의 값을 기준으로 행들을 오름차순(기본값) 또는 내림차순으로 줄 세웁니다.
언제 쓰나요? 매출 데이터에서 '매출액'이 가장 높은 상위 10개 고객을 찾고 싶거나, 시계열 데이터를 날짜 순서대로 깔끔하게 나열하고 싶을 때 사용합니다.

4. DataFrame.reset_index(drop=True)
역할: 데이터프레임의 인덱스(행 번호) 초기화 및 정리
기능: 데이터를 필터링하거나 위의 sort_values()로 정렬하고 나면 기존의 행 번호(0, 1, 2...)가 뒤죽박죽 섞이게 됩니다. 이때 섞인 행 번호를 다시 0부터 순서대로 깔끔하게 매겨줍니다.
drop=True의 의미: 이 옵션을 넣지 않으면 과거의 뒤죽박죽된 인덱스가 표 안에 새로운 열(Column)로 쏙 들어와 버립니다. drop=True는 "옛날 인덱스는 필요 없으니 아예 휴지통에 버려라"라는 뜻입니다.

5. DataFrame.transform()
역할: 데이터 변환 및 원본 형태 유지
기능: 데이터에 특정 연산이나 함수를 적용해 줍니다. 특히 groupby()와 짝꿍으로 쓰일 때 아주 강력합니다.
언제 쓰나요? groupby()로 그룹별 평균을 구하면 결과가 요약되어 표의 길이나 모양이 작아집니다. 하지만 transform()을 쓰면 원본 데이터의 행 개수와 모양을 그대로 유지하면서, 각 행의 옆에 그 행이 속한 그룹의 평균값을 채워 넣어줍니다. 원본 데이터와 그룹 요약 데이터를 1:1로 비교해야 할 때 유용합니다.

6. MinMaxScaler.fit_transform() (Scikit-Learn 라이브러리)
역할: 데이터 정규화(Normalization) / 스케일링
기능: 머신러닝 모델을 돌릴 때, 어떤 변수는 값이 1~10이고 어떤 변수는 값이 1000~10000이면 모델이 큰 숫자에만 영향을 받는 문제가 생깁니다. 이를 방지하기 위해 데이터의 최솟값을 0, 최댓값을 1로 맞추어 모든 데이터를 0~1 사이의 비율로 압축해 줍니다.
fit_transform()의 원리: 데이터의 최솟값과 최댓값을 찾는 학습 과정(fit)과, 그 기준에 맞춰 실제로 숫자들을 0과 1 사이로 바꾸는 과정(transform)을 한 번에 실행해 주는 아주 편리한 함수입니다.
```
