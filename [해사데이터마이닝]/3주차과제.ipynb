```python
[문제8] 결측치가 있는 행을 제거한 후 학생이 가장 많이 수강한 과목을 찾고, 해당 과목 점수를 표준화(스텐더드 스케일)한 다음 가장 큰 표준화된 값을 구하시오. (반올림하여 소수 셋째 자리까지 계산)

정답:
import pandas as pd
# df = pd.read_csv("student_assessment.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch7/student_assessment.csv")

# 1) 결측치 제거
# print(df.shape)
df = df.dropna()
# print(df.shape)

# 2) 가장 많이 수강한 과목 필터링
id = df['id_assessment'].value_counts().idxmax()
cond = df['id_assessment'] == id
df = df[cond]

# 3) 과목 점수 스탠다드 스케일
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df['score'] = scaler.fit_transform(df[['score']])
#scaler의 method 가 2차원인 dataframe 을 입력으로 사용
# 4) 가장 큰 값
print(round(df['score'].max(), 3))

답: 2.183
```

```python
1. DataFrame 메서드 (Pandas 라이브러리)
데이터를 표(엑셀 형태)로 다룰 때 사용하는 핵심 기능들입니다.

1) quantile()
핵심 정의: 데이터의 특정 분위수(Percentile) 값을 구하는 메서드입니다.
주요 특징: 0부터 1 사이의 값을 입력하여 사용합니다. (예: 0.5를 넣으면 중간값, 0.25를 넣으면 하위 25% 값) 데이터의 분포를 파악하거나 이상치(튀는 값)를 찾아낼 때 주로 사용합니다.
쉬운 예시: 학교 시험 점수에서 "상위 10% 커트라인 점수가 몇 점인지" 알아낼 때 사용합니다.

2) sort_values()

핵심 정의: 특정 열(Column)의 값을 기준으로 데이터를 정렬하는 메서드입니다.
주요 특징: 오름차순(기본값) 또는 내림차순(ascending=False)으로 정렬할 수 있습니다.
쉬운 예시: 학생들의 성적 데이터를 '수학 점수'가 높은 사람부터 1등, 2등 순서대로 줄 세울 때 사용합니다.

3) isnull()

핵심 정의: 데이터 안에 결측치(비어있는 값, NaN)가 있는지 확인하는 메서드입니다.
주요 특징: 값이 비어있으면 True, 값이 들어있으면 False를 반환합니다. 보통 isnull().sum() 형태로 연결해서 각 열마다 비어있는 데이터가 총 몇 개인지 세는 데 활용합니다.
쉬운 예시: 설문조사 결과를 모았을 때, '나이' 칸을 빈칸으로 제출한 사람이 있는지 찾아냅니다.

4) dropna()

핵심 정의: 결측치(비어있는 값)가 포함된 행이나 열을 아예 삭제해 버리는 메서드입니다.
주요 특징: 데이터를 가장 빠르고 단순하게 정제하는 방법이지만, 유용한 데이터까지 통째로 날아갈 수 있어 주의해서 써야 합니다.
쉬운 예시: 설문지 중에서 단 한 문항이라도 답을 안 적은 사람의 설문지는 통계에서 아예 빼버리고 버리는 것과 같습니다.

5) value_counts()
핵심 정의: 특정 열에서 각 고유값들이 몇 번씩 등장하는지 개수를 세어주는 메서드입니다.
주요 특징: 주로 범주형 데이터(카테고리)의 분포를 확인할 때 쓰며, 가장 많이 나온 값부터 내림차순으로 보여줍니다.
쉬운 예시: 혈액형 데이터에서 A형 몇 명, B형 몇 명, O형 몇 명, AB형이 몇 명인지 각각 세어서 표로 보여줍니다.

6) max() 와 idxmax() (비교해서 알아두면 좋습니다)
max() 핵심 정의: 데이터에서 가장 큰 값 자체를 반환합니다.
쉬운 예시: "우리 반에서 제일 높은 수학 점수는 100점이야."
idxmax() 핵심 정의: 데이터에서 가장 큰 값을 가진 항목의 **인덱스(위치나 이름)**를 반환합니다.
쉬운 예시: "우리 반에서 수학 점수가 제일 높은 사람은 김철수야."

2. Python 기본 함수
데이터 분석 라이브러리 없이도 파이썬 자체에서 기본적으로 제공하는 필수 함수입니다.

1) abs()
핵심 정의: 숫자의 절댓값(Absolute value)을 반환하는 함수입니다.
주요 특징: 양수는 그대로, 음수는 부호를 떼고 양수로 만듭니다. 오차를 계산하거나 두 값의 순수한 차이(거리)를 구할 때 필수적입니다.
쉬운 예시: 주식이 어제보다 5% 올랐든(+5), 5% 떨어졌든(-5), '변동폭' 자체는 5%라고 계산하고 싶을 때 사용합니다.

2) len()
핵심 정의: 객체(리스트, 문자열, 데이터프레임 등) 안에 들어있는 항목의 총개수(길이)를 반환하는 함수입니다.
주요 특징: 데이터의 전체 크기를 확인할 때 가장 먼저 쳐보는 기본 함수입니다.
쉬운 예시: 책이 총 몇 페이지인지, 혹은 고객 명단에 총 몇 명의 이름이 적혀있는지 세어줍니다.

3. StandardScaler 객체 메서드 (Scikit-learn 라이브러리)
머신러닝을 돌리기 전, 데이터의 스케일(단위)을 맞춰주는 과정에서 쓰입니다.

1) fit_transform()

핵심 정의: 데이터의 패턴을 학습(Fit)하는 동시에, 그 패턴에 맞춰 데이터를 변환(Transform)하는 두 가지 작업을 한 번에 실행하는 메서드입니다.
주요 특징: StandardScaler는 데이터를 평균이 0, 표준편차가 1인 정규분포 형태로 바꿔줍니다(표준화). fit_transform()은 학습용 데이터(Train data)의 평균과 분산을 계산(fit)한 뒤, 즉시 그 기준에 맞춰 데이터의 크기를 조절(transform)합니다.
쉬운 예시: '키(cm)'와 '몸무게(kg)'는 단위도 다르고 숫자 단위의 크기도 다릅니다. 이 둘의 평균을 각각 구해서(Fit), 공평하게 비교할 수 있는 완전히 새로운 점수 체계로 데이터를 싹 바꿔버리는(Transform) 작업을 동시에 수행합니다.
```
