```python
문제 1) 단일표본 t-test: 배달시간 평균 검정
문제 제시문 한 배달 플랫폼은 “서울 A권역의 평균 배달 완료 시간은 30분 이내”라고 홍보하고 있다.운영팀은 최근 1개월간 무작위로 추출한 주문 데이터를 이용하여 이 주장이 통계적으로 타당한지 검정하려고 한다.

모범답안 Q1. 가설 설정 • 귀무가설 (H_0): 서울 A권역의 평균 배달시간은 30분 이상이다. (mu >= 30) • 대립가설 (H_1): 서울 A권역의 평균 배달시간은 30분 미만이다. (mu < 30) Q2. 검정 방법 선택 • 한 집단의 평균을 기준값 30과 비교하므로 단일표본 t-test를 사용한다. Q3. 검정 수행 • ttest_1samp()로 t-statistic과 p-value를 구한다. Q4. 신뢰구간 제시 • 표본평균과 95% 신뢰구간을 계산한다. Q5. 결론 • p-value < 0.05 이고 표본평균이 30보다 작으면 “평균 배달시간이 30분 미만”이라고 판단한다. • 그렇지 않으면 홍보 문구를 통계적으로 뒷받침하기 어렵다.

Python 예시 코드 import pandas as pd from scipy import stats import numpy as np df = pd.read_csv("delivery_time.csv") x = df["delivery_min"].dropna() # 단일표본 t-test (대립가설: 평균 < 30) result = stats.ttest_1samp(x, popmean=30, alternative="less") mean_x = x.mean() n = len(x)

se = stats.sem(x) ci_low, ci_high = stats.t.interval( 0.95, df=n-1, loc=mean_x, scale=se ) print("표본평균:", mean_x) print("t-statistic:", result.statistic) print("p-value:", result.pvalue) print("95% CI:", (ci_low, ci_high))

ttest_1samp ( a , popmean , axis = 0 , nan_policy = 'propagate' , alternative = 'two-sided' , * , keepdims = False )[원천] 한 그룹의 점수 평균에 대한 T-검정을 계산하십시오. 이는 독립적인 관측값 표본 a 의 기대값(평균)이 주어진 모집단 평균 popmean 과 같다는 귀무가설을 검정하는 것입니다 .

핵심 결과:- 표본평균: 27.946분 - t-statistic: -4.1398 - p-value: 0.000090 - 95% 신뢰구간: (26.9424, 28.9496) 결과에 대한 해석:단일표본 t-test를 실시한 결과, 평균 배달시간의 표본평균은 27.946분이었고, 검정통계량은 t=-4.1398, p-value는 0.000090으로 나타났다. p-value가 유의수준 0.05보다 작고 표본평균 또한 기준값 30분보다 작으므로, 평균 배달시간은 30분 미만이라고 통계적으로 해석할 수 있다. 또한 95% 신뢰구간은 26.9424분 ~ 28.9496분으로 나타나 평균 배달시간이 30분보다 낮은 수준에 있음을 뒷받침한다.

최종 해석 유의한 경우:“단일표본 t-test 결과 p-value가 0.05보다 작고 표본평균이 30분보다 작으므로, 서울 A권역의 평균 배달시간은 30분 미만이라고 볼 수 있다.” 유의하지 않은 경우:“단일표본 t-test 결과 p-value가 0.05 이상이므로, 평균 배달시간이 30분 미만이라고 통계적으로 결론내리기 어렵다.” 단일표본 평균 검정은 SciPy의 ttest_1samp로 수행할 수 있습니다.
```

```python
문제 2) 대응표본 t-test: 교육 전후 업무처리시간 비교
문제 제시문 한 기업은 데이터 자동화 교육을 실시한 뒤, 직원들의 보고서 작성 시간이 줄었는지 확인하고자 한다.같은 직원 25명의 교육 전후 평균 작성시간(분)을 수집하였다

모범답안 Q1. 가설 설정 • 귀무가설 (H_0): 교육 전후 평균 업무처리시간 차이는 없다. (mu_d = 0) • 대립가설 (H_1): 교육 후 업무처리시간이 감소했다. (mu_d > 0), 단 (d = before - after) Q2. 검정 방법 선택 • 동일한 직원의 교육 전·후 데이터를 비교하므로 대응표본 t-test를 사용한다. Q3. 검정 수행 • ttest_rel(before, after, alternative="greater")를 사용한다. Q4. 평균 차이 계산 • before - after의 평균을 계산한다. 값이 양수면 평균적으로 시간이 감소한 것이다. Q5. 결론 • p-value < 0.05 이고 평균 차이가 양수면 교육이 업무처리시간 단축에 기여했다고 본다.

Python 예시 코드 import pandas as pd from scipy import stats df = pd.read_csv("before_after_worktime.csv") before = df["before_min"] after = df["after_min"] diff = before - after result = stats.ttest_rel(before, after, alternative="greater")
print("평균 차이(before - after):", diff.mean()) print("t-statistic:", result.statistic) print("p-value:", result.pvalue)

핵심 결과:- 교육 전 평균: 62.5856분 - 교육 후 평균: 56.8844분 - 평균 차이(before - after): 5.7012분 - t-statistic: 14.7794 - p-value: 0.000000 결과에 대한 해석:대응표본 t-test를 실시한 결과, 교육 전 평균 업무처리시간은 62.5856분, 교육 후 평균 업무처리시간은 56.8844분으로 나타났으며, 평균 차이(before-after)는 5.7012분이었다. 검정통계량은 t=14.7794, p-value는 7.49e-14로 유의수준 0.05보다 매우 작으므로, 교육 후 업무처리시간이 유의하게 감소했다고 해석할 수 있다. 따라서 해당 교육은 업무 효율 개선에 통계적으로 유의한 효과가 있다고 볼 수 있다.

최종 해석 유의한 경우:“대응표본 t-test 결과 p-value가 0.05보다 작고 교육 전후 평균 차이(before-after)가 양수이므로, 교육 후 업무처리시간이 유의하게 감소했다고 볼 수 있다.” 유의하지 않은 경우:“대응표본 t-test 결과 p-value가 0.05 이상이므로, 교육 후 업무처리시간이 감소했다고 통계적으로 말하기 어렵다.” 대응표본 비교는 SciPy의 t-test 계열 함수로 처리하는 것이 일반적입니다
```
